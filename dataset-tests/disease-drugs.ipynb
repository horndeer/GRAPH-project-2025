```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset_path = "../datasets/ChSe-Decagon_monopharmacy.csv.gz"

# Read the CSV file with pandas first for better handling of the data
df = pd.read_csv(dataset_path, compression='gzip')
df["Individual Side Effect"].unique()
```

```python
# Create the directed graph from the dataframe
nx_graph = nx.from_pandas_edgelist(df, '# STITCH', 'Individual Side Effect', ['Side Effect Name'], create_using=nx.DiGraph)

# Display basic graph statistics
print(f"Number of nodes: {nx_graph.number_of_nodes()}")
print(f"Number of edges: {nx_graph.number_of_edges()}")
```

```python
subgraph = nx_graph.subgraph(list(nx_graph.nodes)[:2000])
nx.draw(subgraph, node_size=1, with_labels=False, pos=nx.bipartite_layout(subgraph, list(nx_graph.nodes)[:2000]))
```
